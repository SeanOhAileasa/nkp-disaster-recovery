###### 1. Availability Concepts
***

In this presentation we'll take a look at several concepts that are all related to improving the availability of your applications, services, and even hardware. 

Now, some of them might seem a little bit similar and at the end of the day, they probably are but they do have slightly different implementations and/or configurations. 

But, ultimately, they are all concerned with ensuring that you can continue to provide services, even when things fail.

So the first example is what's known as fault tolerance.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Fault Tolerance - RAID
***

And this is the ability to respond to unexpected failure of either hardware or software. 

And the goal is to deliver uninterrupted service in the event of failure of one or more components. 

Now a good example of this is the redundant array of independent disks, or RAID.

Now if you aren't familiar with RAID, it is a collection of physical hard disks that operate as a single logical unit. 

So let's say you had five physical hard disks as far as the operating system is concerned, they are all a single, logical D drive or E drive. 

The letter doesn't really matter but as data is written to the array, it breaks it up into little chunks and spreads it across all disks but it also writes a mathematical algorithm that can be used in conjunction with the existing data to quite literally reconstruct the missing data that you would lose if a disk fails.

And in the higher RAID levels, you can even survive more than one disk failing but again, the idea is that the surviving disks and their associated data, plus these algorithms, can be used to regenerate the missing data, so that when a client requests any given file, they still get their file and as an administrator, you can take out the failed disk, replace it with a new one and then tell the array to re-write that lost data. 

Because again those algorithms quite literally figure out what used to be there so you've got a new disk in place, the data is re-written and everything is exactly as it was. And again the clients would have never known that the disk failed.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### High Availability
***

Now, high availability is similar. 

And those two terms of fault tolerance and high availability will often be interchanged, and that's fine.

But in my experience, anyway, I tend to hear high availability used with reference to software applications, and fault tolerance to hardware but that's by no means rule. 

So high availability is provided by systems or components that can continuously operate without a failure. 

Now, in this example we see multiple copies of a database on separate servers, so again, just a common example. 

But, you typically see one server designated as a primary. And it services all client requests. 

But it also implements replication techniques, so that it maintains a copy of that same database on a separate standby server. 

So if the primary goes down, the secondary is there, ready to assume the workload because it has the exact same copy of the exact same database and while the primary is down, the secondary will store all of the changes that occur and when the primary comes back, it will replicate those changes back to the primary so that they are both in sync with each other and the clients will be transferred back. 

So again, from the client perspective, nothing happened. 

There might have been a minor disruption of service - maybe they had to refresh screen or reconnect, but minimal disruption. 

So again, high availability typically refers to things like that where you have a server that fails and another server comes in to save the day, if you will. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Load Balancing
***

Now, load balancing, in and of itself, is not actually designed to provide fault tolerance - it just inherently does, and I'll clarify that in a moment.

But its primary purpose is to just divide the workload of a single computer across multiple computers in a balanced fashion. 

So that, in this example, Server 1, that is hosting a website, does not become overwhelmed with client requests while Server 2, hosting the same site, hardly gets any requests. 

It distributes the traffic across all of them in a very even fashion.

Can implement a load balancer with either hardware or software. 

But again, load balancing in and of itself doesn't really provide fault tolerance - it just gives you better performance. 

But coming back to the inherent fault tolerance, it's the fact that you now have multiple servers, all performing the same job, all actively servicing client requests. 

That gives you fault tolerance or high availability, because let's say the client was connected to Server 3. 

If Server 3 should go down, you might think, well, the client has lost its connection and therefore can no longer can get to that site but the load balancer will realize that Server 3 went down and it will redirect the client off to a surviving server. 

So again by balancing the traffic it always has to do so across only active servers, so it will never forward a client request to a server that has gone down, so that's where you get the inherent high availability or fault tolerance through load balancing. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### NIC Teaming
***

NIC teaming is quite simply the process of using multiple network adapters - might hear some similar terms, including link bundling, network bonding, or link aggregation. 

But it really is just the use of multiple adapters to aggregate the bandwidth and its primary function is to increase the available bandwidth. 

If you have bandwidth x with one adapter, you have bandwidth 2x with two adapters.

You can quite literally plug them both in and aggregate that bandwidth. 

But again, because you have two physical adapters, you inherently get some fault tolerance. 

If adapter one should fail, or if somebody just inadvertently unplugs it by mistake, adapter two is still there. 

It's rather unlikely that both of them would fail or both get unplugged at exactly the same time. 

Now the whole server can go down of course, but with respect to just the network adapters, you do gain that fault tolerance by having two of them. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Clustering
***

Clustering kind of combines a little bit of everything. 

It is just the act of connecting two or more computers together to ultimately behave as one computer, so that's kind of like fault tolerance. 

But you see it's often used for large parallel processing workloads, load balancing and fault tolerance. 

So we see a little bit of everything and the fact that you have more than one computer kind of looks like high availability.

So to give you an idea of how it's a little bit different, if you had, let's say, ten servers, they would all be performing the same task - that's like load balancing. 

But whereas in high availability, we only had one active server and one or more standby servers, in clustering, you can have several active servers and several standby servers.

Now you can configure clustering in a lot of different ways. 

But again, just to give you that example of using ten servers, you could have five of them active, so with that you get load balancing, and you also get high availability because you have five of them all doing the same job.

But in terms of fault tolerance, should one of those servers go down, you could probably still continue to provide services but you might think, well, you know what, four servers aren't enough to handle the workload. 

So I'm going to take one of the standbys and bring them into the mix so I still have five active servers. 

So again, you get a little bit of everything. 

Now clustering can become a little more complex and it can be more expensive obviously but it is very reliable, and very high performing as well. 

The larger your cluster, the more reliable, and the better it will perform. 

But you don't have to go all crazy with clustering. 

You can implement a cluster with as little as two computers. So ultimately you can configure any of these and/or all of them to improve your overall availability.

###### 2. Power Management
***

Another key consideration with respect to availability, and perhaps the most important consideration, is having power - if you don't have any power, of course, you don't have anything. 

So on this presentation we'll take a look at some considerations for improving availability with respect to power management. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Uninterruptible Power Supply (UPS)
***

Now beginning with battery backups and/or uninterruptible power supplies or a UPS.

This is a device that is, quite simply a large battery. 

And it is used to maintain power in the event of a power outage. 

But one thing most people tend to not realize about UPS devices is that they don't last very long. 

It depends, of course, on how much you are drawing from it. 

But UPS's aren't really designed to provide any kind of extended service. 

If you think the power's going to be out for several hours or even days, they are really more so designed to just give you time to gracefully shut everything down. 

Most high level components in IT don't like just instantly losing power and immediately just shutting off. 

They want to go through their shut down process, so they give you maybe 15 minutes, maybe half an hour, maybe even a couple hours with the larger, more robust models. 

But the idea is to just be able to go around, take advantage of that time, and shut everything down correctly and gracefully. 

That way, when the power does come back, everything has gone through its appropriate processing. 

You might even want to actually physically unplug some of your more delicate devices, if you will because if a surge happens, then they can still be damaged, so you wait until everything comes back up, then you restart everything. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Power Management
***

Now, at the larger scales, you might find a power generator and these are certainly still used to power devices in the event of an outage. 

But typically you find these in the larger environments and used when there are larger scale or more widespread outages that could last for days because these run on fuel. 

So as such, they certainly can be a lot more expensive and a lot more complex because they do actually run wires from them to certain devices in a much more permanent fashion. 

So it may not be entirely reliable because you might not always know what can be powered by the generator and of course, you might still run out of fuel. 

So it is certainly better than a battery, but certainly much more expensive as well. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Dual Power Supply
***

Now, at a much smaller or more focused scale, you can implement dual power supplies. 

Now, this is the component that supplies power to your computer and if you have sort of a custom desktop system, you might be familiar with the power supply and they're all rated in terms of watts, but more powerful computers, quite simply, need power supplies with more watts. 

So if you have a very robust desktop system, or perhaps maybe a custom-built server that isn't a rack-mount type of device, then you might just have more devices within that system and as such, they will need more power. 

For example, you might have a raid array with five or ten physical hard drives. 

Maybe you have multiple optical drives. 

There are probably a lot of fans in the housing because it gets hotter. 

So, as such, you need more power, so you can, in several instances in the larger systems, mount more than one power supply to give you that additional power. 

And that does lend a little bit of fault tolerance as well because if one of them does fail, you could still maybe plug the critical components into the surviving power supply and continue to provide services. 

But again, it's just simply due to the fact that you have a lot of physical devices in those systems and as such, you simply need more power. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Circuit
***

And finally, the circuit, or more to the point, redundant circuits doesn't necessarily refer to power management. 

But it is a configuration to possibly mitigate the potential for isolated outages, maybe within a building or within a data center.

If, for example, there is some construction going on and something just gets broken or mistakenly taken down - might just lose power in an isolated area. 

Maybe this rack of switches or those servers in that server room, so by implementing redundant circuits or multiple physical paths for traffic to travel, you might be able to, for lack of a better word, sidestep that outage. 

So we see redundant switches and redundant routers and different ways to get to the WAN. 

So that if switch A is the one that went down, then perhaps we can also go through switch B, and the same with the routers. 

So again, this is just a way to possibly mitigate these isolated outages by implementing these redundant paths, so it´s not so much a power management technique, if you will, but rather a way to mitigate the potential for an isolated power outage.

So ultimately, everything that you can do to help ensure that you still have power when you lose power, will clearly help to improve your availability of service.

###### 3. Recovery
***

In this presentation, we'll take a look at some considerations for recovery. 

But not really data recovery, but rather an entire location, a physical site, an office building or an entire data center. 

Because, of course, things do happen that can take down large geographic regions. 

Hurricanes, floods, earthquakes, things along those lines can have, of course, a serious impact on physical locations. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Recovery – Hot Site - Recovery Time Objective (RTO)
***

So we'll begin with what's known as the hot site and this is typically something you see implemented at the commercial levels, the higher scale enterprise businesses. 

And it is quite simply a location that is entirely set up and ready to go in the event of a failure of an entire data center or an office building and it more or less is an exact replica, at least to the point where it is fully equipped. 

There is office space, there's furniture, there's phone jacks. 

There are people to staff it and of course, there is computer equipment. 

So everything is there and ready to go with a hot site configuration.

So this is designed to pretty much allow for continuous operation in the event of a disaster and has what's known as a RTO, or a recovery time objective, of hours. 

> In other words, the only delay is how long does it take to get your people from the down site to the hot site? 

So in some cases, that can be very little time. 

The hot site itself is constantly updated. 

t's maintained both with hardware and software and in some cases, you might even see certain operations taking place in the hot site.

And a good example is, in fact, backups - might replicate your data to the hot site and then perform your backups there.

But however it's used is your call, but one of the significant considerations here, of course, is the cost to maintain a hot site. 

But if your business is deemed to be exceptionally critical in terms of needing to provide service, then a hot site is a good solution. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Recovery – Warm Site
***

A warm site is still additional office space or data center space that is available and there might be some preinstalled hardware and there would be servers that are in a ready state to become production environment servers. 

But in a lot of cases, the components are typically shared among multiple customers. 

So in fact, they are essentially saying, well, you can have some of the space, not all of it and then these people have some of the space and those people have some of the space, so you do see a little bit of a shared environment. 

But it is ideal for businesses that aren't deemed to be as critical. 

But you still need some level of redundancy, so there's basic infrastructure there, but it's not ready to go at a moment's notice. 

So the recovery time objective with the warm site might be measured in days. 

It would rely on your existing backups for recovery, so there's no data ready to go and it would typically implement some kind of shared storage, which can be less expensive because, again, other people are sharing the cost with you, but you might not have as much capacity. 

Now, with respect to the backups themselves, disk-based backups are recommended over tape in this case because this can improve the recovery time.

Disks are quite simply faster, so you can get up and running a little more quickly. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Recovery – Cold Site
***

Finally, the cold site is certainly still used in disaster recovery situations, but it really is nothing more than the physical space itself. 

So if, for example, you have just an empty warehouse basically, you might use that as your cold site. 

There's no hardware, there's no data, there's no people to staff it basically. 

So the customer provides everything that is necessary, all of the equipment. 

They have to perform all of the installations required for continuous operations, and basically set everything up themselves. 

Now, the recovery time with a cold site is typically measured in weeks or more, possibly even months or years, depending on the nature of the disaster. 

If, for example, it was an earthquake and the entire building has collapsed, then you certainly aren't going to be up and running again in a few days or a week. 

You may have to entirely relocate for a very extended period of time. 

So in most cases, cold sites are really only implemented for businesses that can be down for an extended period of time but sometimes they're used to complement options in the event of a disaster, meaning that you might in fact have a hot site and a cold site. 

So the hot site would effectively be plan A.

If a disaster happens, then we move to the hot site but maybe the hot site was taken down as well. 

So then plan B is the cold site, and you may not even have one ready to go. 

It could be a matter of finding a cold site first, but at least maybe it's in the budget or it has been considered, so you might only see this in larger organizations, but as mentioned, it's kind of the plan B scenario. 

Maybe you might have to go quite far away before you find a viable cold site. 

But again, it's all about trying to maintain service throughout the event of any kind of disaster, so whatever you can implement is certainly going to help keep you up and running as effectively as possible.

###### 4. Backups
***

In this presentation, we will take a look at what is perhaps the most important aspect of IT administration, performing backups.

Now, in our previous presentation, we talked about restoring, but that was at an entire site level - for this presentation, we are talking about data. 

So ensuring that you have multiple copies of your data, so that you can recover anything that might be lost. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Full
***

So we'll begin with the full backup - this is certainly the most complete option available. 

It makes a complete copy of all data that you tell it to, so you say back up these folders, those drives, it doesn't matter. 

Everything you specify will be backed up, usually daily, so whatever you specify is backed up every day. 

It copies that data to another set of media, including tape, disk or DVD. 

That's your call, and you can use a combination of them but again, what it basically does is to simply back up everything, regardless of whether or not that file has already been backed up and/or if the file has changed. 

So even if you backed it up yesterday and then it did not change the next day, it will still be backed up. 

So it's just very safe, you basically aren't taking any chances. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Full - Advantages
***

Now, in terms of advantages, this offers a quicker recovery time objective, or RTO. 

> It is quite simply how long will it take to recover? 

So with a full, you can just say, well, restore either everything that was lost on that disk or restore just this folder, in case of an accidental deletion or things like that. 

So it can be very quick to recover because you just say, well, there's the backup, put it back, we're back in business. 

Its very simple. 

Again, you just don't have to concern yourself with whether or not any given file has changed. 

You just say I don't care, back it up anyway, so it's very safe. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Full - Disadvantages
***

But in terms of disadvantages, it certainly can take longer to perform a full backup, because of the fact that you are saying back up everything. 

And of course, it requires more space, so as new data comes along after the full, on Day-1, Day-2, and Day-3, any new data just gets piled in to the full backup, so they certainly can tend to grow. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Incremental
***

The next option is what's known as the incremental. 

And this backs up only the data that has changed since the last full backup or the last incremental. 

So this tends to make a little more sense. 

You might think, well, if I create a file, let's say, on Monday, and then I back it up and it does not change, then there's no point in backing up that file on Tuesday. 

The original backup that happened on Monday is perfectly fine - the data did not change. 

But when files do change, they create this little timestamp that effectively says I have changed since the backup and the backup process picks up on that and says all right, you have changed. 

Therefore, I will back you up again but if you did not change, there's no point in backing you up because you've already been backed up. 

So again, each day's incremental backup simply looks back to the previous day and said:

> Did you change? 

If you did, I'll back you up - if you did not, you've already been backed up. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Incremental - Advantages
***

So the advantages with the incremental basically come down to the fact that you can have a faster backup. 

Because you are backing up far less data for easy math let's just say if on Day-1, only 10% of the data has changed since the full, the previous day, call it Day-0. 

Then it's only 10% as large as that original backup. 

So it's faster and requires less space. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Incremental - Disadvantages
***

But it does have a disadvantage on the recovery side. 

It can take longer to recover everything because you have to restore the original full backup and then every incremental up to the point of failure to get everything back up and running. 

So again, you have to restore the full but that does not take into account all the changes that happened on Day-1, 2, and 3. 

So then you have to restore Day-1, then Day-2, then Day-3, and you have to go in that sequence to get all those changes put back, so it can take longer to recover everything. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Differential
***

Now the differential is similar, in that it backs up only files that have been modified but it's always since the last full.

So where an incremental always looks back to the previous day, regardless of whether that was a full or an incremental, the differential always looks back to the full, so always back to Day-0, if you will. 

So subsequent differential backups will continue to include all data since the last full. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Differential – Advantages
***

And this has the advantage of requiring less storage space compared to a full but when you perform the recovery, it offers a faster recovery than, say, an incremental because it only requires the full as the baseline and the last differential. 

Because it continues to capture all changes since the last full. 

So if the disaster happened after Day-2, we can restore the full that was on Day-0. 

Then we don't have to worry about Day-1 or Day-2.

We can just go to Day-3 and restore that differential because that has all changes gathered up since the full. 

So it's a faster restore than the incremental, but still has some disadvantages. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Differential – Disadvantages
***

The backup process itself is slower than the incremental. 

It still requires the full as a baseline and it might take longer to restore specific files. 

If you're just looking for this particular file, you might not be sure. 

> Is it in the full, or is it in Day-3? 

But it all comes down to what you feel is more important to you. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Backups – Options
***

Do you need faster backups, or do you need faster restores? 

Unfortunately, you kind of have to sacrifice one for the other, so you have to consider how long it takes to perform the backup. 

If it's running all night and into the morning when people are showing up to do their work and the backup is still happening, that's going to interfere, so if the faster backup is a little more important, then maybe you look at the incremental. 

But in most environments, when a disaster happens, it's get us back up and running as quickly as possible. 

So maybe the differential is a little more important, so you always have to consider a number of different options. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Backups – Snapshots
***

And finally, the snapshot is used to establish recovery points. 

And this is not unlike within your operating system. 

Many operating systems establish these recovery points, which are quite literally a picture of the state of everything at that point in time. 

These can be easier and faster than traditional backup options and offer very fast recovery time and recovery point objectives. 

They can restore individual lost files or even recover from substantial data loss. 

But the idea is that it literally just takes a picture of the state of that data, so that you can say, well, all right, looks like things have gone wrong here but I know everything was fine yesterday at 2 PM, so let's just go back to that point and then continue on from there. 

But again, you do have to consider the amount of data in any given snapshot. 

So if, for example, let's just go with something fairly small, if the snapshot is 1 GB of data, and you take 1 per hour or every hour of the business day. 

Let's just say 10 hours, then you have 10 GB worth of snapshots that can pile up fairly quickly. 

So it clearly depends on the size of the snapshot and the frequency but they are very fast for just saying, let's just go back three hours, ten hours to 2 PM yesterday. 

Then you basically say, that's what I want to use, let's continue on from that point. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Backups in Place in all Situations
***

One way or another, it is particularly important to ensure that you have backups made and again, it really comes down to what you feel is the best approach and you can use combinations of them, but you do have to make sure that you have backups in place in all situations.

###### 5. Mean Time to Failure (MTTF)
***

Now, in this presentation, we'll take a look at a value that is known as the MTTF, or mean time to failure. 

This is the length of time a certain product is expected to last in operation, and this is based on specific and rigorous testing of that product. 

But at the end of the day, we are talking about hardware and in a lot of cases where you will see this value, it refers to hardware that is generally non-repairable. 

A very good example of this is a hard drive - generally don't send a hard drive in for repair, you just replace it. 

It really just doesn't make sense to repair it. 

The cost would be, in most cases, more than what it would cost to buy a new one. 

So generally, we are talking about things like hard drives, but more generally, things that tend to have moving parts, so there are certainly a lot of other components such as fans and coolant systems that will also sooner or later wear out. 

So it generally is not found on something like a switch, for example - there really aren't any moving parts in that type of hardware. 

But anything that does really tend to wear out would have a mean time to failure value. 

Now, this is common when testing equipment that is deemed to be mission critical and again, things like a hard drive, where they store your critical data is certainly deemed to be mission critical, so this is of great interest to engineers when they are trying to improve upon their devices and it is, of course, valuable for determining the overall reliability of components or entire systems from the perspective of those who are purchasing and administering that equipment. 

So again, a hard drive is just a very good example of this value being implemented and you might see something like an MTTF of around a million hours, so you can more or less expect the devices to start failing once they reach that amount of time. 

Now, how quickly they will actually wear out, of course, depends on how heavily they are used, so devices that are much more frequently accessed are more likely to fail sooner than those that aren't accessed as much but the million hours, in this case, indicates a million hours of usage.

So that, of course, can vary with respect to how much actual calendar time passes because, again, if you aren't using it very often, maybe it's a dedicated backup device and is only written to for a few hours a day, well, then it's going to take a lot longer for it to reach 1 million hours of actual use. 

But it is something to consider, particularly when you are buying new equipment or considering upgrades or things along those lines. 

Hardware will eventually fail, so you ideally want to get as much life out of that hardware as possible before it does fail, so it is, again, something that you would want to consider when making these upgrades, these new purchases, or perhaps even starting from scratch. 

So the longer the MTTF, the better the value will be.

<a id="dissasterRecovery6MeanTimeBetweenFailuresMTBF"></a>
***
###### 6. Mean Time Between Failures (MTBF)
***

Another value that is used with respect to hardware or device failure - MTBF, or the mean time between failures.

In our previous presentation, we talked about the mean time to failure, or MTTF, and that one is typically used in reference to components that aren't repaired - you just replace the entire component. 

The MTBF typically refers to components that can be repaired. 

Now, there's a little bit of crossover, if you will, in that regard, as a simple example, consider a power supply. 

A power supply is a fairly replaceable component, as is, but in a lot of cases it is the fan that fails for the power supply, not the entire power supply. 

It will usually outlive the fan by quite a significant degree. 

So in this regard, you can consider the power supply as being repairable by just replacing the fan.

So that's a simple example, but the same might apply to a high level video adapter card.

They tend to need a lot of cooling as well, so, again, you can just replace the fan, and that is really considered to be a repair of the entire unit. 

So, again, you expect these devices to fail, but we can implement some type of repair. 

So it still measures the hardware reliability in terms of any given product or component and it's typically measured in hours. 

For example, a component could be measured in thousands or even tens of thousands of hours between the failures. 

But, again, the idea is that you can just replace or repair that single component and the entire unit continues to function just fine.

###### Quantifiable Objective Measure
***

So this is often used as a quantifiable, objective measure when designing new products and the results can be derived from intensive testing, product experience, and analyzing known factors that might affect the lifespan of any given product. 

But with respect to what it means to you as a consumer, it helps you to know what kind of service plan might be necessary to go along with this device.

To give you a more common day to day example, if you think about your car, you buy the car, and obviously there are things, with respect to the entire car, that can be repaired and we expect things to fail at certain points. 

We have to replace the brake pads every so often. 

We have to change the oil and the filter. 

So this is, of course, the service plan that goes along with the car, so, there is a similar approach to the components and the devices that you purchase in the world of IT as well. 

Certain things will have to be serviced, so we can, in a lot of cases, purchase some kind of service plan and then, we know that we can rely on the fact that the manufacturer or some sort of representative will show up and perform these regular repairs, so that we can keep things in order and hopefully avoid the point where everything just does start to fail to a significant degree. 

So it's just upkeep, it's maintenance, and knowing the MTBF can help you to design a better, more effective service plan to keep things running as effectively and as efficiently as possible.
